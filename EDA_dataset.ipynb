```python
speaker_id = []

for text_json_path in tqdm(text_json_lst):
    # with open(text_json_path) as f:
    #     sample_json = json.load(f)
    speaker_id += list(x.replace('speakerjj', '') for x in text_json_path.split('/')[-1].split('_') if 'speaker' in x)
```

```python
value_counter = Counter(speaker_id)
value_counter
```

```python
train_id, valid_id = train_test_split(speaker_id, test_size=0.2, random_state=42)
```

```python
train_info = {'path': [], 'dialect': [], 'standard': []}
valid_info = {'path': [], 'dialect': [], 'standard': []}

for text_json_path in tqdm(text_json_lst):
    # with open(text_json_path) as f:
    #     sample_json = json.load(f)
    speaker = list(x.replace('speakerjj', '') for x in text_json_path.split('/')[-1].split('_') if 'speaker' in x)
    
    with open(text_json_path) as f:
        sample_json = json.load(f)
    
    dialect = ' '.join(list(x['dialect'] for x in sample_json['transcription']['segments']))
    standard = ' '.join(list(x['dialect'] if x['standard'] is None else x['standard'] for x in sample_json['transcription']['segments']))
    
    if speaker[0] in train_id:
        train_info['path'].append(text_json_path)
        train_info['dialect'].append(dialect)
        train_info['standard'].append(standard)
    else:
        valid_info['path'].append(text_json_path)
        valid_info['dialect'].append(dialect)
        valid_info['standard'].append(standard)
```
